```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
x
```

```python
y=3*x
y
```

```python
plt.plot(y,x,color='g',linestyle=':',linewidth='5')
plt.title("X Versus Y")
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.show()
```

```python
x=np.arange(1,11)
y1=2*x
y1
y2=3*x
y2
```

```python
plt.plot(x,y1,color='b',linestyle='-',label='dotted')
plt.plot(x,y2,color='g',linestyle=':',label='lined')
plt.title("two line in 1 plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

```python
x=np.arange(1,11)
y1=2*x
y1
y2=3*x
y2
```

```python
plt.subplot(2,1,1)
plt.title("two line in 1 chart")
plt.grid(True)
plt.xlabel("x-axis")
plt.plot(x,y1)
plt.ylabel("y-axis")
plt.subplot(2,1,2)
plt.plot(x,y2,color='r')
plt.grid(True)
plt.show()
```

```python
stu={'bob':70,'mat':78,'sam':98}
names=list(stu.keys())
values=list(stu.values())
```

```python
plt.bar(names,values,color='r')
plt.title("bar chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.grid(True)
plt.show()
```

```python
plt.barh(names,values,color='b')
plt.title("bar chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.grid(True)
plt.show()
```
